```python
#Example data from previous competition

import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

#matplotlib.style.use('ggplot')
sns.set()
df = pd.read_csv('000_data_v2.csv')
df.columns = [col.lower() for col in df.columns]
df.sort_values(by='obc_timestamp', inplace=True)
df.reset_index(drop=True, inplace=True)
df.fillna(method='ffill',inplace=True)
df.head(10)
```

```python
import numpy as np

listVel = [3.6, 7.2, 10.8, 14.4, 18, 21.6, 25.2, 28.8, 32.4, 36, 39.6, 43.2]
potencias = [348, 617, 813, 965, 1080, 1075, 817, 488, 328, 296, 188, 150]
df.gps_speed.values 
```

```python
newVel = []
potDf = []
for value in df.gps_speed.values:
    subList = abs(listVel - value)
    idx = np.argmin(subList)
    newVel.append(listVel[idx])
    potDf.append(potencias[idx])
    
df['newVels'] = newVel
df['potencias'] = potDf
df.head(10)
```

```python
P = df['potencias']
df['energy2'] = np.cumsum(P*0.1/1000)# kJ 
df['energy2_km_kWh'] = df['distance']/(df['energy2']/3600) # Km/kWh
df.head(10)
```

```python
fig, ax1 = plt.subplots(1,1, figsize = (22,14))

paleta = sns.color_palette("dark:#5A9_r", as_cmap=True)
paleta = sns.color_palette("rocket", as_cmap=True)
paleta = sns.color_palette("mako", as_cmap=True)
paleta = sns.color_palette("magma", as_cmap=True)


fontdictTitle = {'fontsize': 40,
 'fontweight' : 4,
 'verticalalignment': 'baseline',
 'horizontalalignment': 'center'}

fontdictSubTitleY = {'fontsize': 26,
 'fontweight' : 2,
 'verticalalignment': 'bottom',
 'horizontalalignment': 'center'}

fontdictSubTitleX = {'fontsize': 26,
 'fontweight' : 2,
 'verticalalignment': 'top',
 'horizontalalignment': 'center'}

sns.lineplot(ax=ax1, data=df, x='timeMin', y='energy2', hue='lap', palette=paleta, linewidth=4)
ax1.set_title('Energy vs time', fontdict=fontdictTitle)
ax1.set_xlabel('Time (s)', fontdict=fontdictSubTitleX)
ax1.set_ylabel('Energy (kJ)', fontdict=fontdictSubTitleY)
ax1.tick_params(axis='both', which='major', labelsize=24)
ax1.get_legend().remove()

# add legend
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='right', ncols=1, bbox_to_anchor=(0.98, 0.75),fontsize=24)

fig.savefig('./energy2.png', dpi=300)

plt.show()
```
